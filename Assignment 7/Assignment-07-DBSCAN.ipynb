```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler
plt.rcParams['figure.figsize']=(10,7)
plt.rcParams['figure.dpi']=250
%matplotlib inline
```

```python
df=pd.read_csv('crime_data.csv')
df
```

```python
df.info()
```

```python
data=df.drop(['Unnamed: 0'],axis=1)
data
```

```python
# Feature scaling
```

```python
sc=StandardScaler()
x=sc.fit_transform(data)
x
```

```python
#  Hierarchical clustering
```

```python
#   Dendrogram with single and complete linkage
```

```python
from scipy.cluster import hierarchy
```

```python
lk=hierarchy.linkage(x,method='complete')
dendo=hierarchy.dendrogram(lk)
plt.tight_layout()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# Create Clusters (y)
hc=AgglomerativeClustering(n_clusters=4)
hc
```

```python
y=hc.fit_predict(data)
y
```

```python
data.head()
```

```python
data['cluster']=y
```

```python
#  Analyzing cluster
```

```python
data
```

```python
data.groupby('cluster').agg('mean')
```

```python
plt.scatter(data['cluster'],data['Murder'])
```

```python
# DBSCan
```

```python
x
```

```python
#    MODEL BUILDING
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=1,min_samples=4)
y=db.fit_predict(x)
y
```

```python
data['clusters']=y
data
```

```python
data.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.scatter(data['clusters'],data['UrbanPop'])
```
