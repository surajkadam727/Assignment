#    Calculate VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df.head()
```

```python
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(df.values,i) for i in range (len(df.columns)) ]
vif
```

```python
vif['Features']=df.columns
vif
```
