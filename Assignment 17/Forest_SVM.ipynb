```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
data["month"] =lb.fit_transform(data["month"])
data["day"] = lb.fit_transform(data["day"])
data["size_category"] = lb.fit_transform(data["size_category"])
```

```python
data
```

```python
X = data.iloc[:,:11]
X
```

```python
y=data["size_category"]
y
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y, test_size = 0.3)
```

```python
Xtrain.shape ,Xtest.shape,ytrain.shape,ytest.shape
```

```python
model = SVC()
```

```python
model.fit(Xtrain, ytrain)
```

```python
ypred = model.predict(Xtest)
ypred
```

```python
print(confusion_matrix(ytest, ypred))
```

```python
print(classification_report(ytest, ypred))
```

Train model accuracy is 0.77

```python
model.score(Xtest,ytest)
```

Model is not predicting well, so we will improve the model by hyperparameter tunning using grid search method.

```python
#  Improving Model using Grid Search CV
```

```python
param_grid = {'C' : [0.1, 1, 10, 100, 1000], 'gamma' : [1, 0.1, 0.01, 0.001, 0.0001], 'kernel' : ['rbf']}
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3,  cv = 5)
```

```python
grid.fit(Xtrain, ytrain)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_pred = grid.predict(Xtest)
```

```python
grid_pred
```

```python
print(classification_report(ytest, grid_pred))
```

So, using grid search method, we improved our model accuracy to 0.92
